````python
# Formatando nossa API em Json:
url_json = r.json()
url_json
````

````python
# Gerando um DataFrame com nosso Json:
df_url = pd.DataFrame(url_json['items'])
df_url
````

````python
# Criando filtros para que nosso DataFrame receba um valor específico para as colunas 'UF' e 'interessado':
filtro_uf = df_url['uf'] == 'RS'
filtro_interessado = df_url['interessado'] == 'Tabaí'

# Aplicando os filtros criados:
df_url_filtrado = df_url[filtro_interessado & filtro_uf]
df_url_filtrado
````

````python
# 2) 

# Criando filtros para que nosso DataFrame receba um valor específico para as colunas 'UF' e 'interessado':
filtro_uf_mg = df_url['uf'] == 'MG'
filtro_status_2 = df_url['status'] == 'Arquivado por decurso de prazo'
````

````python
# Aplicando os filtros criados:
df_url_filtrado_2 = df_url[filtro_status_2 & filtro_uf_mg]
````

````python
# Exibindo as dimensões do nosso DataFrame, onde as linhas são equivalentes a quantidade de solicitações para o Estado:
df_url_filtrado_2.shape
````

````python
# 3) 

# Criando filtros para que nosso DataFrame receba um valor específico para as colunas 'UF' (onde nos trará apenas resultados referentes a Bahia) e 'tipo_interessado' (para que possamos filtrar apenas os munícipios e não considerar os campos preenchidos como "estado"):
filtro_uf_ba = df_url['uf'] == 'BA'
filtro_tipo_municipio = df_url['tipo_interessado'] == 'Município'

# Aplicando os filtros ao DataFrame original e solicitando a contagem de cada valor diferente do campo que informa as cidades:
df_url[filtro_uf_ba & filtro_tipo_municipio].value_counts('interessado')
````

````python
# 4) 

# Criando filtros para que nosso DataFrame receba um valor específico para a coluna 'tipo_interessado' (para que possamos filtrar apenas os munícipios e não considerar os campos preenchidos como "município"):
filtro_tipo_estado = df_url['tipo_interessado'] == 'Estado'
````
